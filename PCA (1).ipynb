```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv('/content/wine.csv')
wine
```

```python
wine['Type'].value_counts()
```

```python
wine2=wine.iloc[:,1:]
wine2
```

```python
wine2.shape
```

```python
wine2.info()
```

```python
# Converting data to numpy array
wine_ary=wine2.values
wine_ary
```

```python
# Normalizing the numerical data
wine_norm=scale(wine_ary)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color='magenta')
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:2],columns=['PC1','PC2',])],axis=1)
final_df
```

```python
# Visualization of PCA
sns.scatterplot(final_df)
```
